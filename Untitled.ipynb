Заметим что тип prior_question_had_explanation object, хотя должен быть по логике Boolean. Исправим эту неточность

```python
# Преобразование NaN в False и изменение типа на boolean
train_df['prior_question_had_explanation'] = train_df['prior_question_had_explanation'].fillna(False).astype(bool)

# Проверка результата
print(train_df['prior_question_had_explanation'].info())
print(train_df['prior_question_had_explanation'].value_counts())
```

Также в user_answer и в answered_correctly в train встречаются -1 (min в describe), возможно это указание на nan объекты 

```python
train_df["user_answer"].count()
```

```python
# Заменить -1 на NaN только в столбце user_answer
train_df["user_answer"].replace(-1, pd.NA, inplace=True)

# Проверка результата
print(train_df["user_answer"].value_counts(dropna=False))
```

```python
# Преобразование NaN в False и изменение типа на boolean
train_df['answered_correctly'] = train_df['answered_correctly'].fillna(False).astype(bool)

# Проверка результата
print(train_df['answered_correctly'].info())
print(train_df['answered_correctly'].value_counts())
```

```python
print("Content Type Distribution:")
print(train_df['content_type_id'].value_counts())

# Анализ answered_correctly по content_type_id
plt.figure(figsize=(8, 5))
sns.barplot(x='content_type_id', y='answered_correctly', data=train_df)
plt.title('Average Correct Answers by Content Type')
plt.xlabel('Content Type')
plt.ylabel('Average Correct Answers')
plt.xticks([0, 1], ['Question', 'Lecture'])
plt.show()
print(2)
```
